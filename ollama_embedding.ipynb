# Ollama
- ollama support embedding models, making it possible to build retrieval augmented generation(REG) applications that combine text promts with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b") ## by default it uses llama2
)
```

```python
embeddings
```

```python
# List of text wnat to embedded
r1=embeddings.embed_documents([ "Alpha is the first letter of the greek alphabet" ,
                               "beta is the second letter of the greek alphabet",
                               ])
```

```python
len(r1[0])  # number of dimesion.
```
