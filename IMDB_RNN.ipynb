```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense,Flatten
```

```python
# pip install --upgrade certifi
```

```python
## Load the imdb datasets
max_features=10000 # vocabulary size
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

import tensorflow as tf

# Set up to ignore SSL verification (for development purposes only)
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# Load dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=max_features)


print(f"Training data shape: {x_train.shape}, Training labels shape: {y_train.shape}")
print(f"Testing data shape: {x_test.shape}, Testing labels shape: {y_test.shape}")
```

```python
# INspect a sample review and its label
sample_review=x_train[0]
sample_label=y_train[0]
```

```python
print(f"Sample review {sample_review}")
print(f"Sample label :{sample_label}")
```

```python
# Mapping of words index back to index (for understanding)
word_index=imdb.get_word_index()
# word_index
```

```python
max_len=500
x_train=sequence.pad_sequences(x_train, maxlen=max_len)
x_test=sequence.pad_sequences(x_test, maxlen=max_len)
print(x_train[0])
```

```python
model=Sequential()
model.add(Embedding(10000, 128,input_length=max_len)) # vocubalary, features_dimesions, max_input_length-length of senteces.
model.add(SimpleRNN(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
## Create an instance of Earlystopping callback
from tensorflow.keras.callbacks import EarlyStopping
early_stopping=EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
early_stopping
```

```python
# Train the model with early stoppiing
model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    callbacks=[early_stopping],
    epochs=10,
    batch_size=32
)
```

```python
model.save('simple_rnn_imdb.h5')
```

```python
word_index=imdb.get_word_index()
```

```python
reverse_word_index={value:key for key, value in word_index.items()}
```

```python
# reverse_word_index
```

```python
from tensorflow.keras.models import load_model
```

```python
model=load_model('simple_rnn_imdb.h5')
```

```python
model.summary()
```

```python
# model.get_weights()
```

```python
# Helper function
def decode_review(encoded_review):
  return ' '.join([reverse_word_index.get(i-3 , '?') for i in encoded_review])

def perprocess_text(text):
  words=text.lower().split()
  encoded_review=[word_index.get(word,2)+3 for word in words]
  padding_review=sequence.pad_sequences([encoded_review], maxlen=500)
  return padding_review
```

```python
# word_index
```

```python
sample_review=x_train[0]
decode_review(sample_review)
```

```python
def predict_sentiment(review):
  preprocessed_text=perprocess_text(review)
  prediction=model.predict(preprocessed_text)
  sentiment='positive' if prediction[0][0] > 0.5 else 'negative'
  return sentiment, prediction
```

```python
sentiment, prediction=predict_sentiment("This movies was fantastic. The acting was great and the plot was thrilling")
print(f"Movies review: {sentiment}")
print(f"Prediction score: {prediction}")
```
