## Embedding Technique
- COnverting text into vectors

```python
from langchain_openai import OpenAIEmbeddings
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
embeddings
```

```python
text="This is a tutorial on openAI embedding"
quary_result=embeddings.embed_query(text)
```

```python
len(quary_result) # by default given text in converted into 3072 dimension.
```

```python
# Embedding with defining dimenstion
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
text="This is a tutorial on openAI embedding"
quary_result=embeddings_1024.embed_query(text)
quary_result
```

```python
len(quary_result)
```

```python
from langchain_community.document_loaders import TextLoader
```

```python
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_spliter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
```

```python
final_documents=text_spliter.split_documents(docs)
final_documents
```

## Now converting text into Embedding and store into vectorBD database
- Vector Embedding and Vector StoreDB

```python
from langchain_community.vectorstores import Chroma
```

```python
db=Chroma.from_documents(final_documents, embeddings_1024)
```

```python
db
```

```python
## Retrieve the results from quary vectorstore db
quary="It will be all the"
retrieved_results=db.similarity_search(quary)
print(retrieved_results)
```
